# list question

```python
# find the largest num in a list

marks=[56,99,76,72,88,64]
largest_number=marks[0]
for i in marks:
    if i > largest_number:
        largest_number=i
print(largest_number)

```

```python
# find the second largest number
marks=[56,99,76,72,88,64]
# sorted_marks=sorted(marks)
# sorted_marks[-2]
first=second=marks[0]
for i in marks:
    if i > first:
        second=first # 56=99
        first=i
    elif i > second and i!=first:
        second=i
print(second)
```

```python
#remove duplicaata from list
num=[1,23,41,1,52,1,2,8]
result=[]
for i in num:
    print(i)
    if i  not in result:
        result.append(i)
print(result)
```

```python
#move zeores to end
move_zero=[0,1,0,2,4]
result=[]
for i in move_zero:
    if i != 0:
        result.append(i)
print(result)
for i in move_zero:
    if i==0:
        result.append(i)
print(result)
```

# dict problems


```python
# count frequency of each number
num=[1,3,4,2,5,8,9,3,1,2,3,9,8,3,2,1]
frequency={}
for i in num:

    if i not in frequency:
        frequency[i]=1
    else:
        frequency[i]+=1
print(frequency)
```

```python
# count frequency of words
fruit=["apple","banana","carrot","apple","carrot","carrot","banana"]
frequency={}
for i in fruit:
    if i not in frequency:
        frequency[i]=1
    else:
        frequency[i]+=1
print(frequency)
```

```python
# find the first mark of the student
marks={
    "naveen":99,
    "ravi":85,
    "hari":77,
    "nithish":89

}
top_student=""
top_mark=0


for k,v in marks.items():
   if v > top_mark:
      top_student=k
      top_mark=v
print(top_mark)
print(top_student)
```
